```python
def clean_about(df):
    final_list=[]
    about=df.About
    
    for row in about:
        try: 
            split_text=row.split("\n")
            split_text.remove("About")
        except: 
            split_text=None
        final_list.append(" ".join(split_text))
    return final_list
```

# OrganizationInformation

```python
# Lista de tags posibles, incluyendo Financing
tags = ['Scholarship type', 'Number of scholarships to award', 'Grant', 'Scholarship coverage', 'travel expenses',
        'Applicable programmes', 'Benefits', 'Financing']

# Función para extraer bloques de texto entre un tag y el siguiente tag en la lista
def extract_section(tag, text, tags=tags):
    # Crear un patrón para capturar texto entre el tag actual y el siguiente en la lista de tags
    next_tag_pattern = '|'.join(tags)  # Unir todos los tags con "OR" para usar en la búsqueda
    pattern = rf"{tag}\n((?:[^\n]+\n)+?)(?={next_tag_pattern}|$)"
    match = re.search(pattern, text)
    if match:
        return match.group(1).strip()
    return None
```

```python
def OrganisationInformation(df):
    org_info = df['OrganisationInformation']
    
    # Inicializar las listas para cada atributo
    list_scholarship_type = []
    list_num_scholarships = []
    list_grant = []
    list_scholarship_coverage = []
    list_Description = []
    list_applicable_programmes = []
    list_benefits = []
    list_Financing = []

    # Lista de tags a considerar
    tags = ['Scholarship type', 'Number of scholarships to award', 'Grant', 'Scholarship coverage', 'Description', 
            'Applicable programmes', 'Benefits', 'Financing']
    
    # Iterar sobre cada texto en 'OrganisationInformation'
    for text in org_info:
        list_scholarship_type.append(extract_section('Scholarship type', text))
        list_num_scholarships.append(extract_section('Number of scholarships to award', text))
        list_grant.append(extract_section('Grant', text))
        list_scholarship_coverage.append(extract_section('Scholarship coverage', text))
        list_Description.append(extract_section('Description', text))
        list_applicable_programmes.append(extract_section('Applicable programmes', text))
        list_benefits.append(extract_section('Benefits', text))
        
        # Para Financing, capturar todo lo que queda hasta el final
        financing_text = re.search(r'Financing\n(.*)', text, re.DOTALL)
        if financing_text:
            list_Financing.append(financing_text.group(1).strip())
        else:
            list_Financing.append("Not found")
    
    # Agregar las nuevas columnas al DataFrame al final
    df = df.assign(
        Scholarship_Type=list_scholarship_type,
        Number_of_Scholarships=list_num_scholarships,
        Grant=list_grant,
        Scholarship_Coverage=list_scholarship_coverage,
        Description=list_Description,
        Applicable_Programmes=list_applicable_programmes,
        Benefits=list_benefits,
        Financing=list_Financing
    )

    return df  # Devolver el DataFrame modificado
```

```python
def Eligibility(df):
    Eligibility = df['Overview']
    print(Eligibility)
    # Inicializar las listas para cada atributo
    list_Eligibility = []
    list_Disciplines = []
    list_Locations = []
    list_Nationality = []
    list_study_experience_required = []
    list_age = []

    # Lista de tags a considerar
    tags_2 = ['Eligibility','Scholarship requirements','Disciplines', 'Locations', 'Nationality', 'Study experience required', 
            'Age', 'Duolingo English Test']
    
    # Iterar sobre cada texto en 'Eligibility'
    for text in Eligibility:
        list_Eligibility.append(extract_section('Eligibility', text, tags_2))
        list_Disciplines.append(extract_section('Disciplines', text, tags_2))
        list_Locations.append(extract_section('Locations', text, tags_2))
        list_Nationality.append(extract_section('Nationality', text, tags_2))
        list_study_experience_required.append(extract_section('Study experience required', text, tags_2))
        list_age.append(extract_section('Age', text, tags_2))
        
    
    # Agregar las nuevas columnas al DataFrame al final
    df = df.assign(
        Eligibility_crit=list_Eligibility,
        Disciplines=list_Disciplines,
        Locations=list_Locations,
        Nationality=list_Nationality,
        study_experience_required=list_study_experience_required,
        age=list_age
    )

    return df  # Devolver el DataFrame modificado
```
